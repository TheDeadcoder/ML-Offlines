```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data_path = 'data/train.csv'
test_data_path = 'data/test.csv'

train_data = pd.read_csv(train_data_path)
test_data = pd.read_csv(test_data_path)

print("Number of training examples: ", train_data.shape[0])
print("Number of test examples: ", test_data.shape[0])
```

## Basic Exploratory data analysis

**Number of Features**

```python
print("Number of features: ", train_data.shape[1])
```

**Displaying Sample Images**

```python
class_labels = {
    0: "T-shirt/top",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle boot"
}

def plot_sample_images(data, num_images=25):
    plt.figure(figsize=(10,10))
    for i in range(num_images):
        plt.subplot(5,5,i+1)
        image = data.iloc[i, 1:].values.reshape(28,28)
        plt.imshow(image, cmap='gray')
        plt.title(class_labels[data.iloc[i,0]])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Plotting 25 sample images
plot_sample_images(train_data, num_images=25)
```

**Class Distribution**

```python
def plot_class_distribution(data, title='Class Distribution'):
    plt.figure(figsize=(8,6))
    sns.countplot(x=data['label'], palette='viridis')
    plt.title(title)
    plt.xlabel('Class Label')
    plt.ylabel('Count')
    plt.xticks(ticks=range(10), labels=[class_labels[i] for i in range(10)], rotation=45)
    plt.show()

plot_class_distribution(train_data, title='Training Data Class Distribution')

plot_class_distribution(test_data, title='Test Data Class Distribution')
```
